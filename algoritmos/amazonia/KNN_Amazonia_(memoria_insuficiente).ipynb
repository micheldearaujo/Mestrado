# Aqui implementaremos o algoritmo 'K-Nearest Neighbors' para classificar imagens do dataset Amazonia

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Flatten, Dropout, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import SGD
from sklearn.metrics import fbeta_score as fbeta, confusion_matrix as cm, classification_report as cr
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from datetime import timedelta
import time
```

```python
# Definindo o caminho dos diretorios
base_dir = 'D:/michel/data/amazonia/kaggle'
train_dir = os.path.join(base_dir, 'train-jpg')
test_dir = os.path.join(base_dir, 'test-jpg')
train_fnames = os.listdir(train_dir)
test_fnames = os.listdir(test_dir)
```

```python
# Para carregar os dados
data = np.load(base_dir +'/amazon_data_128.npz')
X, y = data['arr_0'], data['arr_1']
print('Dimensões: ')
print('X: ',X.shape, '\n y: ', y.shape)
```

```python
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Verificando a dimensão dos dados. São 60 mil imagens no total, onde 50 são de treinamento e 10mil para teste
print(Xtr.shape)
print(ytr.shape)
print(Xte.shape)
print(yte.shape)
```

```python
Xtr = Xtr.reshape(Xtr.shape[0], 128*128*3) ## Vamos concatenar os dados das 3 dimensoes em apenas 1 dimensão
Xte = Xte.reshape(Xte.shape[0], 128*128*3)
```

```python
print(Xtr.shape)
print(ytr.shape)
print(Xte.shape)
print(yte.shape)
```

Vamos dividir o set de testing para conseguir um set de validação, pela metade:

```python
Xval = Xte[:6072,:]
yval = yte[:6072]
```

```python
print(Xval.shape)
print(yval.shape)
```

```python
Xte = Xte[6072:,:]
yte = yte[6072:]
```

```python
print(Xte.shape)
print(yte.shape)
```

# Treinando o modelo
Agora vamos criar nosso modelo de K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Escolhendo o melhro valor para k: Lembrando que vamos usar primeiro o set de validação !!
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
import time
from datetime import timedelta
start_time = time.monotonic()

error_rate = []

# Will take some time
for i in range(1,2):
    print('loop %s'%i)
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(Xtr,ytr)
    pred_i = knn.predict(Xval)
    error_rate.append(np.mean(pred_i != yval))
    
end_time = time.monotonic()
print(timedelta(seconds=end_time - start_time))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

# Conclusão:

O K-nearest neighbors é um algoritmo de Machine Learning de classificação que funciona da seguinte forma: Dado um conjunto de dados de treinamento, ele pega cada dado de teste e compara com cada um dos dados de treinamento, procurando os dados mais próximos. Dando um valor de K, ele irá classificar o dado de teste baseado nos K dados mais próximos.
No nosso exemplo do CIFAR-10, usamos apenas 10% dos dados e obtivemos uma acurária de 40% para K = 10 e 34% para K=1, com um tempo de cálculo de aproximadamente 40 minutos.
