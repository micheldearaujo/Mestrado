# Aqui implementaremos o algoritmo 'K-Nearest Neighbors' para classificar imagens do dataset CIFAR-10

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Getting the data
```

```python
from Dados.data_utils import load_CIFAR10
Xtr, Ytr, Xte, Yte = load_CIFAR10('Dados/cifar-10-batches-py')
```

```python
# Verificando a dimensão dos dados. São 60 mil imagens no total, onde 50 são de treinamento e 10mil para teste
print(Xtr.shape)
print(Ytr.shape)
print(Xte.shape)
print(Yte.shape)
```

```python
Xtr = Xtr.reshape(Xtr.shape[0], 32*32*3) ## Vamos concatenar os dados das 3 dimensoes em apenas 1 dimensão
Xte = Xte.reshape(Xte.shape[0], 32*32*3)
```

```python
print(Xtr.shape)
print(Ytr.shape)
print(Xte.shape)
print(Yte.shape)
```

Vamos dividir o set de testing para conseguir um set de validação, pela metade:

```python
Xval = Xte[:5000,:]
Yval = Yte[:5000]
```

```python
print(Xval.shape)
print(Yval.shape)
```

```python
Xte = Xte[5000:,:]
Yte = Yte[5000:]
```

```python
print(Xte.shape)
print(Yte.shape)
```

# Treinando o modelo
Agora vamos criar nosso modelo de K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Escolhendo o melhro valor para k: Lembrando que vamos usar primeiro o set de validação !!
```

```python
import time
from datetime import timedelta
start_time = time.monotonic()

error_rate = []

# Will take some time
for i in range(1,40):
    print('loop %s'%i)
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(Xtr,Ytr)
    pred_i = knn.predict(Xval)
    error_rate.append(np.mean(pred_i != Yval))
    
end_time = time.monotonic()
print(timedelta(seconds=end_time - start_time))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

# Classificação Final
Agora vamos classificar de fato novas imagens e ver a acurácia do nosso modelo.

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(Xtr, Ytr)
```

```python
pred = knn.predict(Xte)
```

```python
# Evaluating
```

```python
from sklearn.metrics import classification_report as cr, confusion_matrix as cm
```

```python
print('Classification Report:')
print(cr(Yte, pred))
print('\n')
print('Confusion Matrix:')
print(cm(Yte, pred))
```

# Conclusão:

O K-nearest neighbors é um algoritmo de Machine Learning de classificação que funciona da seguinte forma: Dado um conjunto de dados de treinamento, ele pega cada dado de teste e compara com cada um dos dados de treinamento, procurando os dados mais próximos. Dando um valor de K, ele irá classificar o dado de teste baseado nos K dados mais próximos.
No nosso exemplo do CIFAR-10, usamos apenas 10% dos dados e obtivemos uma acurária de 40% para K = 10 e 34% para K=1, com um tempo de cálculo de aproximadamente 40 minutos.
